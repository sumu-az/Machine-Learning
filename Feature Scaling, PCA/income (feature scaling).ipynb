### Read data from income.csv and scale the relevant features using standardization

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

data_set = pd.read_csv('income.csv')
data_set
```

```python
x = data_set[['Age', 'Income']]
x
```

### MIN MAX SCALER (Normalization)

```python
from sklearn.preprocessing import MinMaxScaler

MMS = MinMaxScaler()

x_after_min_max_scaler = MMS.fit_transform(x)

x_after_min_max_scaler
```

### Standardisation

```python
from sklearn.preprocessing import StandardScaler

Standardisation = StandardScaler()

x_after_Standardisation = Standardisation.fit_transform(x)

x_after_Standardisation
```
